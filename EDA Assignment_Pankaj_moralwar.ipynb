## 5. Creating def functions

#### We will create some pre defined functions for our easyness

```python
def create_pie_plot(Column,dataframe):
    #lables = dataframe[Column].value_counts()
    lables = list(pd.unique(dataframe[Column]))
    plt.figure(figsize = [6,7])

    #dataframe[Column].value_counts()
    plt.pie(dataframe[Column].value_counts(), labels=lables, autopct='%1.1f%%', 
            textprops={'fontsize': 15, 'fontweight' : 15, 'color' : 'Black'}, startangle=90,labeldistance=1)
    #plt.legend
    plt.legend(bbox_to_anchor=(1, 0, 0.5, 1), loc="center left", fontsize=12, bbox_transform=plt.gcf().transFigure)
   
    title = "Distribution of "+ Column.title() + " Variable"
    plt.title(title, fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})

    plt.show()
```

```python
def create_pie_plot2(Column,dataframe):
    plt.figure(figsize=[15,7])
    dataframe[Column].value_counts(normalize=True).plot.pie(autopct='%1.1f%%',textprops={'fontsize': 8, 'fontweight' : 8, 'color' : 'Black'},labeldistance=1.05)
    title = "Distribution of "+ Column.title() + " Variable"
    plt.title(title, fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})
    plt.ylabel("")
    plt.show()
```

```python
def create_barh_plot(Column,dataframe):
    plt.figure(figsize = (15,20))
    dataframe[Column].value_counts().plot.barh()
    title = "Distribution of "+ Column.title() + " Variable"
    plt.title(title,color="Red", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'blue'})
    plt.xticks(size=15,color='Black')
    plt.show()
```

```python
def create_bar_plot(Column,dataframe):
    dataframe[Column].value_counts().plot.bar()
    title = "Distribution of "+ Column.title() + " Variable"
    plt.title(title,color="Red", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'blue'})
    plt.xticks(size=15,color='Black')
    plt.show()
```

```python
def create_dist_plot(Column,dataframe):
    sns.set_style("dark")
    sns.distplot(dataframe.Column,bins=20,color="Green")
    title = "Distribution of "+ Column.title() + " Variable"
    plt.title(title,color="Green", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})
    plt.show()
```

```python
def create_hist_plot(Column,dataframe):
    dataframe[Column].value_counts().plot.hist()
    title = "Distribution of "+ Column.title() + " Variable"
    plt.title(title,color="Green", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})
    plt.show()
```

```python
# function to count plot for categorical variables
def univariate_categorical(var):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=var, data=def_df,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(def_df))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=var, data=ndef_df,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=15)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(ndef_df))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```

```python
# function to dist plot for continuous variables
def create_unidist_plot(var):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15,5))
    
    sns.distplot(a=def_df[var],ax=ax1)

    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15)
            
    sns.distplot(a=ndef_df[var],ax=ax2)
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=15)    
        
    plt.show()
```
